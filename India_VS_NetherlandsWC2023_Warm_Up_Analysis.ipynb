```python
# Program for Random Input

import matplotlib.pyplot as plt
from random import randint


overs = list(range(1, 51))  # Create a list of overs from 1 to 50
score_India = [randint(1, 10) for _ in range(50)]  # Generate random scores for India
score_Netherlands = [randint(1, 10) for _ in range(50)]  # Generate random scores for the Netherlands


total_India = sum(score_India)  # Calculate the total score for India
total_Netherlands = sum(score_Netherlands)  # Calculate the total score for the Netherlands


print("Total Score of India after end of 50 overs:", total_India)
print("Total Score of Netherlands after end of 50 overs:", total_Netherlands)


if total_India > total_Netherlands:
  print("India Won the Match")
elif total_India < total_Netherlands:
  print("Netherlands Won the Match")
elif total_India == total_Netherlands:
  print("Maths has been a Tie")


# Lists to store the cumulative scores
cumulative_score_India = [sum(score_India[:i+1]) for i in range(50)]
cumulative_score_Netherlands = [sum(score_Netherlands[:i+1]) for i in range(50)]


plt.plot(overs, score_India, label="India", color='b')
plt.plot(overs, score_Netherlands, label="Netherlands", color='r')
plt.xlabel("Overs")
plt.ylabel("Score")
plt.title("Cricket Score Comparison")
plt.legend()
plt.show()


plt.bar(overs, score_India, label="India", color='b')
plt.bar(overs, score_Netherlands, label="Netherlands", color='r')
plt.xlabel("Overs")
plt.ylabel("Score")
plt.title("Cricket Score Comparison")
plt.legend()
plt.show()


plt.bar(overs, score_India, label="India", color='b')
# plt.bar(overs, score_Netherlands, label="Netherlands", color='r')
plt.xlabel("Overs")
plt.ylabel("Score")
plt.title("Cricket Score Comparison")
plt.legend()
plt.show()


# plt.bar(overs, score_India, label="India", color='b')
plt.bar(overs, score_Netherlands, label="Netherlands", color='r')
plt.xlabel("Overs")
plt.ylabel("Score")
plt.title("Cricket Score Comparison")
plt.legend()
plt.show()


plt.plot(overs, cumulative_score_India, label="India", color='b')
plt.plot(overs, cumulative_score_Netherlands, label="Netherlands", color='r')
plt.xlabel("Overs")
plt.ylabel("Cumulative Score")
plt.title("Cumulative Score Comparison")
plt.legend()
plt.show()

```
